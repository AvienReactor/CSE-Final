```python
goodsleepy.loc[goodsleepy['Smoking status'] == 'No', 'Smoking status'] = 0
goodsleepy.loc[goodsleepy['Smoking status'] == 'Yes', 'Smoking status'] = 1
goodsleepy
```

```python
X = goodsleepy[['Sleep efficiency']].values.reshape(-1,1)
y = goodsleepy[['Smoking status']].values.reshape(-1, 1).astype(int)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
# Logistic regression predicting diagnosis from tumor radius
logisticModel = LogisticRegression()
pred = logisticModel.fit(X, np.ravel(y.astype(int)))

# Graph logistic regression probabilities
plt.scatter(X, y)
xDelta = np.linspace(-1, 3, 10000)
yPredicted = logisticModel.predict(X).reshape(-1, 1).astype(int)
yDeltaProb = logisticModel.predict_proba(xDelta.reshape(-1, 1))[:, 1]
plt.plot(xDelta, yDeltaProb, color='red')
plt.xlabel('Sleep Efficiency', fontsize=14)
plt.ylabel('Probability of being Smoker', fontsize=14)
```

```python
pHatProb = logisticModel.predict_proba([[0.3]])
pHatProb
```

```python
probs = pred.predict(X)
```

```python
# labels = np.array(['No']*1250)
# labels[pHatProb >0.5] = "Up"
labels = np.array([0]*388)
labels[probs >0.5] = 1
labels
```

```python
# import ISLP
# from ISLP import confusion_table
# confusion_table(labels, goodsleepy[['Smoking status']])
from sklearn import metrics
metrics.confusion_matrix(y, labels)
# np.mean(labels == goodsleepy[['Smoking status']])
```

```python
Accuracy = metrics.accuracy_score(y, labels)
Accuracy
```

```python
from sklearn.model_selection import train_test_split
# Set the proportions of the training-validation-test split
trainingProportion = 0.70
validationProportion = 0.10
testProportion = 0.20

# Split off the test data
trainingAndValidationData, testData = train_test_split(
    goodsleepy, test_size=testProportion
)

# Split the remaining into training and validation data
trainingData, validationData = train_test_split(
    trainingAndValidationData,
    train_size=trainingProportion / (trainingProportion + validationProportion),
)
```

```python
# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=123
)
```
